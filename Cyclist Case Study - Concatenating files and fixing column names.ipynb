# 0.0 Import

```python
# Importing pandas library

import pandas as pd
```

# 1.0 Handling the data

```python
# Importing csv files from quarters 1 to 4 from 2019

q1 = pd.read_csv('Divvy_Trips_2019_Q1.csv')
q2 = pd.read_csv('Divvy_Trips_2019_Q2.csv')
q3 = pd.read_csv('Divvy_Trips_2019_Q3.csv')
q4 = pd.read_csv('Divvy_Trips_2019_Q4.csv')
```

```python
# Looking at the data

q1.head()
```

```python
# Checking columns names


print("Columns from 2019 1st quarter,", q1.columns)
print("Columns from 2019 2nd quarter,", q2.columns)
print("Columns from 2019 3rd quarter,", q3.columns)
print("Columns from 2019 4th quarter,", q4.columns)
```

```python
# Fixing different column names on the Q2 file


q2 = pd.DataFrame(data=q2.values, columns=q1.columns)
```

```python
# Concatenating files

trips_19 = pd.concat([q1, q2, q3, q4], ignore_index=True)

trips_19
```

```python
# Exporting csv

trips_19.to_csv('divvy_trips_19.csv')
```
